## Problem: BA1B
Find the Most Frequent Words in a String
#### link: https://rosalind.info/problems/ba1b/

```python
def frequent_words(text, k):
    freq = {}
    for i in range(len(text) - k + 1):
        pattern = text[i:i+k]
        if pattern in freq:
            freq[pattern] += 1
        else:
            freq[pattern] = 1
    max_count = max(freq.values())
    k_mers = []
    for pattern in freq:
        if freq[pattern] == max_count:
            k_mers.append(pattern)
    return k_mers


# paste your input here
text = "CTTGAGATCCGGCAATTCGGAGCTTAAGGGAGCTTAAGCGGCAATTCGTCGTAAATCGGAGCTTAAGCTTGAGATCGTGGGAACCCGTGGGAACCCGTCGTAAATCCTTGAGATCCGGCAATTCCGGCAATTCGGAGCTTAAGGTGGGAACCCGTCGTAAATCCGGCAATTCCTTGAGATCCGGCAATTCGTCGTAAATCGTCGTAAATCCTTGAGATCGTCGTAAATCGTCGTAAATCGGAGCTTAAGCTTGAGATCGTGGGAACCCCGGCAATTCCGGCAATTCGGAGCTTAAGGTGGGAACCCCGGCAATTCCGGCAATTCCTTGAGATCGTCGTAAATCCTTGAGATCGTCGTAAATCGGAGCTTAAGGTGGGAACCCGTCGTAAATCGGAGCTTAAGCGGCAATTCGTCGTAAATCGGAGCTTAAGGGAGCTTAAGGTGGGAACCCGTGGGAACCCGTGGGAACCCCTTGAGATCGTCGTAAATCGGAGCTTAAGCTTGAGATCGGAGCTTAAGCGGCAATTCCTTGAGATCCTTGAGATCGTGGGAACCCGGAGCTTAAGGTCGTAAATCGTCGTAAATCGTCGTAAATCGTGGGAACCCGTGGGAACCCGTGGGAACCCGTCGTAAATCGTGGGAACCCGGAGCTTAAGCGGCAATTCGTCGTAAATCCGGCAATTCGTGGGAACCCCTTGAGATCGGAGCTTAAGCGGCAATTCCTTGAGATCGGAGCTTAAGCTTGAGATCGTCGTAAATCCGGCAATTCGGAGCTTAAGGGAGCTTAAGCGGCAATTCGTGGGAACCCGGAGCTTAAGGGAGCTTAAGGTGGGAACCCGGAGCTTAAGCGGCAATTCGTCGTAAATCGGAGCTTAAGCGGCAATTCCTTGAGATCGGAGCTTAAGGGAGCTTAAG"
k = 12
frequent_words(text, k)
```

## Problem: BA1C
Find the Reverse Complement of a String
#### link: https://rosalind.info/problems/ba1c/

```python
def complement(string):
    result = ""
    for i in string:
        if i == 'A':
            result += 'T'
        elif i == 'T':
            result += 'A'
        elif i == 'C':
            result += 'G'
        elif i == 'G':
            result += 'C'
    return result[::-1]


# paste your input here
string = "CTGCTTATGTAGCAAACCGCTGATAGCACCTGAAATCTATGAGCAGAATCTGGTCCGTGTGATCTTCCAGGGAGTCCGTGTGGTAGTTACCCTACGCTCGCATCGAAAGAGAAGAATTTTCGTTAATCCTTGCTCGGCGGTGCCAAGTTACGAACCCGGCCGGTTACCGCCTAGGGCGGTTTCCAAATTATATGGAATGACGAGCATATCGTAAATTGGAGTTGCGCAAAGACCTTGACAGACCGTTAAGCCGAAGTCCGGCCTTTAGTGAGCCCGACTGAGTCAGATGTCAGCTCCGCTAGTAATGAAAACACCTCGGCACCGTCGATCCAACCTGGATAAAGGAGAGTGAACCCGAGGAAGAGGACGCGTAGCAAATGTTTGGTTGCGGGGGGTAGCACAGTACGTACTGTTTCGAGTTCTAGTTTCGGGGTGGATGCCCACTTAGGCTACTCTCTTGGTACCCGTGCTGAATTGTGTAACCCCCCGATCATCATAAGGGCTATTTGAACTGAATCCGTGGTAATATTGAAAGCCTATAGTCCCTAATTGGATGGATCCCCGGGACGCGCGATTTGTACCGACCGCGGATCACATCAGTTAACGCATGATCCGGGAAGGTACGCTAAAATAAATCGTATGGCCTTTGTAGGGCTCGGCCATTGAGTCCGAGATCTAGTATCCGGCTACGGATATAGATGCCAAACTGGCTTCAGCCTATTAGCGCGCTATAAATGCGGCGCCGAACGGGGGAAAGACAGCGCAGTTGCACAATCAAAATCTTCAGTAGATCGGTGGGTAGAACAATTCGCGGCTTAATCTTCCAGAAGGCGGGGGCCCACCTAACGTCCGTCGCAAAAACGTAATGATACGACCCACCAACTCGGCAGTGATGTTGTACTTTAGGCATAATGAAAATTTCCCGGATTGAATCCTTTAAAGATAGACGGCGGGGATGGTCCACCACATCGCAGACGTAATTTCATATCTCCAGTCACGACTCCATCTCGAGCTCTGGAGCTCTTGTAGTCAATGGTGTTCCGACGAAATCATTTCCAGCCGGATCTGCCGCGAACTACATTCTGCGCACCCAGGGTGTCTTGATTCTTACATCGGTCACAGTGGTGTTCGTCATGCATTTGTCTTTTATTTAACAACGGTAGCAGTTTTCGGTGATGACGACTTTACCGTCTTTGTAAGACCTAAATCGATGTCTGCCAGCTCCTGAGTGCCGGGGAACCTGGAAAGAACCCTGCAGGCCGCCGTGATACTGCTCCGATGTGGACGGAGCTTGTAGTGAAAATCAATCTCCTGCAAAGCCGGCGTACTGACCTAAGCACCCGCGAAGCCCCTTGCAGAGCTCTGGATGCCGCGACGGATTTCTTGCCCGGTTTCCGGGAACTATTCCCGGATGACCTAATGGCGGTACGAACGATCAGATAGTGAGGTGAATCACGGGTAAATTATAGTGTTGTAAACAATAGAGCTATACGATGCGATGTACGGTTATTCGCAGCCATCTAAGCCTTAAATCGACAAGTCGATACCTCCTTCTCCCGACCGGACTCACTTGTGCCAACCGATGCTCGGATTTATAGGAACTACGGGTGGGACACTTCGATGATCCGCCCCTATTAAGAATTTCGACCGCATTGATGTACCGACTTCGCTCGGACTACACCGACTCAATGGATCGGACACTGAATTCTTAAACTTCGCTTATTAGATAAAGGCCATGTCGAAGACTGGACTCACATATAGCCGCGACGGCCTGGTTTTCATGTCGCTAAGTAAGTCCGCGATGTCTCAGGTGGATAGAAAATACGCTGGCGCCCGCTCCGACCCGGTATCGTTGACAGTTAGGTTGCCACGGGTGGAGGAATCCGGCGGGGCGTTGTACGCTTAACATATGATACAAGCCTATCTCCTATTCTTCTACCTATTAATCTTAGCGGGTGTCAAAAAAAGGATTGACTGGGGGCGCTTAAGTAGTTTTTAGCTCTTATCCTCGCATTATACCCGATTGCATTTGTTGTCCCGCAATCCCTGAATACTGCGTGCGCAGTTTACACTTTCTGACCATTTTAAGTAGAAGATCGCTACGCATCATTCTTCTCCCGGTATAATATCGGTCGTGGTGTGGGCGACAATTTTTTCATAATTAGTCGGGGTATTACCGTCAAACGCAATCCCAGTCACCGCTGCCCCGCGTCGTCAGGCAGAATTATCGACATTAAGTAGGAGGTGTCGGGGCCGACCTGATGTTACGTCAATGTTAGCTGCACAGGGGTCCCGACGGCGCTGGCGGGCGGGCAACCGGCTGGCGCTACATATGTGGTGCCAGGATCGCAAAAAGCGTCCGTTTCGGGAAACCGTTAAACTCGGTTCCTGACCACGGGGACCTACTGTCTACGTCAACAGCGTACCATTTACCCCCGCATCCTTATCTAGAAGGTTCAAGATACCATGCGTACATCAATACTAGAAGAACCTCGGCGCTGCGACTTTCCGGACCGGTAAGACGCACCGATCTCTTTAACACAGCGCTCGAATCAACTAGAGTTCGGTCCAGCTAGTACCTTCTCGCACGTTGCGGCGTTGATGGGGGTCGTTGGGCGCTTCTCGTGCCAATGTAACACTTGTTTACTTCTTCAGTACTTAGATAGCGGGGAAGTCACCCGGAGTACTGCACGGGCTAACACCGATCAGGACCCTACTACTTGCACGCACGTACTGGCTATTCAACCATCAGCTCTACCCTCGATCCTAACTGGAATATATACCACAGATTTTTGCCAAGTGACATACATGACAGGAAATCCACAAGCTAAGATGCTACGAAATACAAGGCTAGGCAAATGCATAGGCGGTTTTTCATTCTGACATTCTCCATAGTGTGCTGGCGAGACGAAGAAACATCAACTGAATGACCTGACGATCTATGCATGTGTCACAACACCGCAACGAGGTGTGTCTAGCACTATGCAGAAGTCGGGTCCAATGAGGGAAAACGACCACGTATCGCTGCCACGGGACCAGATGATATACTGTGTGAACGCGCGGACAGTTAGGCGTTGCGACGCACATAGAGTCTGGTCACCAGGGGTACTTGCAAGAGCTAGTGAAATGATCACGCAAAATCACTCTAACCGTCAATCGTAAGGTGTTGACCGCACCGCCTAGGGACTAGCCTCTATAACATGTTGTCATAAGTAGGACTCGTCAGCCACGACTAGAGATGATAACGATGCTGTCAGCGTTCTCCGACTAACACACCTATGTCTCAACTAGTTTGTCAAGAGGCACCGAGGCGACGGCCTCCAAGCCGCGATTGTGCCTGTGACAGCCCAGGGACAGAAAATCGCCCTAATAGTCGAGGATGGAAGGCACGGTTGGCTGAATAACACCAATTGTTTATTTCGGTGAAATGGGACTTTGTCACCCCTCGGCGTCCTTGATCTGTTAGGTCCGATTCGCTACTACCGCGTTACCTAACCAAAATGCCCCCTTCGTCGCTACCGAATGACCGCGTGTTGGAAATTTGTCCCGTGTGGACGTTAGCAAAACCCACTACCTGAGTCCTTGATGGCGCCAAGCTTATGGGGTCGACCTCTGCCCCACAATTGCCATCCGCCTCAACACATATCGGGCCACAGATTAGTTCGCCTTAAACCACATCCAATAGGAGCCTGGAGATATCTCGGATCAGTGCAAGTCCGGTATTGTATAAGGTCACACAGCTAGGCCTGGCATCATACTCGCTACAGGTGGGGTCGATAAACGATTTATCGCACCCTGGTTCGGGTTTATAGGTCGCGATGTTCATAGTCATGCATGGAGCACGTAGCGCGTACTCACTTTAAGGGCGTCCCACGGGAAGTCGCTGAACGGAGTCTCCTTGGGCTCTGACTTTAGGCCGTCCTATCACAATATGCCACGCAGGTGTATCACTTACGCCGCAGTAGCTGATCTCCTTACCTTCTCACCCCTGTGGTGGTATAGGCTGGACGTTTCTCTGTCAAACCGCGGCCGCATGTACGCTCCAGTGTGGCGGGCGGGTACATCGGTACGCACCTCGTAGAGAGATATTGGGATTGCTTCAATGTCCTTCCAGCCGCACTGCGACAAGCAATATGGGCGATGACTTCATTTGTATGCGTTGGGTTTGAGAATGTGTAGTACTGGGGATTTCAGCTTCCTTATAGTACGTCGGCGAAGGATGGATTAAACTCTGAAATTGCGAGTATCGCTGCACGCGCGTATGAATGTTTGGAAACGTCTGGATTAATCCTTTGACATCGCGGCGTAGGAAGTTTGCCCTGTTCACGGGCTAGCCTTACAGATAATCTTTCAATGGATCTCCACATGTCGATTAGTTGTAATGAAGATTATTACGACAAATCACCCGAAGTGAAAGATCAATCACCCAACATCGCAACGTATGAGACGGGAGGTCACACCTATAATGCCAGACACTCATAATCTTAAGACTCTTGCAGCATAGTGAGTCTGTGAACGGGCCAACCGTTGTAGCTAGAATTGAGCCGTCGCCCATGTCGTATGTGTAAGCGTAAGGTAACGATTATGGGTAGTGACGCTGTTTCATGGCGTTTACCTCACCCATTTATCAAACACCTAGCAAAATTTAGGAAGGTGACTCATTGAAAAGGTGATATTCCAAAGGTCGCGGAGACTAGAGTACACCTAAGTGGGTAGCAATGAACAGGGGTGAAGTGTGAGATGAACCGTATGCCCCTTTGATCTAACTATCCAGGTCCACCACTGCCATAAGACAGACTTGGGCCCAGTACCTGTAGACACTACAGTCTTTGGTCCCTTGTGTCTGCCGATTGGTTGTGTACTAATATCCCCGCCTTGCAGGCAGGAGCATGAGAGTGGATATATCCCGGTGAGAGATATTGTAACCGTGGGCCGGCTTTGTTCTCAATCCTTATGCAATCGCGGATAGGCGGGCGCATACCGATGACTGTCGAATCGTTACGGGTAACAAGCTCTGTCGAAGCGGGCAACATCCTTACAGTACGCGCGCGACCTTATGACGCCGTTCGGGACTGGCTTCACAAACAGGACTACGGTGGATTCTTACGCGAGTAACAGACCGTCGGCGAAGAGCACGAGGACTTCTATGTCCCGCCGCGGGTCTGTGCCGCTCGGGAGCAGTCCTCCCCAATTGCCTGGGCACGGCATCATCAGCACTTGCCCTCCTATATTGACCGCGAATGCTAGAGGCGTCTTGCAGCCTAGTGTTTCATTTTCCGGCCGTAATTACCCATCGATTGAATGCGCAAGGACAGCACTTACACCGGTTAGTTCACATCAACATTAAGATGGCACACTTCCCCCACGGCGACCATAGCTTGTCGCGAGAATCCGCCAAACCTCCTGGAAGGGGGATGACCCGTTCATCTCGGATGAAACTCGTGCCATAGCGACTCCATAAGCCCGCTGTTTAGTGATTTTTTGACAACGCGGTTGTTACACATTAAAGCGCACCGCCACGTCAATACCGATCCTTCATGCCGAGTCACCGGTCGCACAGCGGATTATGGCTTGCTCAAGTAGGGGGAAACTTCCTTACTTGGGCACTACTACCTTGTACAGGCAAATCCGCTTTGTAGTGGAATCGAGTATACTATGATGACACGATGACAGAACCTCAGTTTACCTGGTCCGCAATCAGCCGATAAGGTAACCAGCTAGTAGTCCGTCTGGCTAATACTTGTTGGTATATATCAGACAAGAATGTGGCTGAAAAGGCACTTTACCTGATTACAATATGTTAGTCCGATTGCTACAACCAATCTAATGCCGAATTGTCAAGTCCGTCTAGCTTTCGACTATGCGTTTAGAGGACTCTAAGGAATTGTCATTATCACGTAGTAAACCGCCTATTCCGTATATATATAGCCCAGGGTCGTTTGCGAATCCCCCATGATGCTACACATAGAGCTCTACAATCACGAGTTTACATGGAGTGAAAGTACCCCGACCAAGGGTAGCTCGCTGTCCAATGGCGGTGCTGGAAATGCCACTCAAGTTTCATTACGGCTAATAGCCAGATGGTGTGACCGGATCGGTGGTCCCAAATATGGTACGATAAATCACGAGCCAAGAGCGTCGGCCGCGCCTGCGTGAAATGTATTCCCAGACGGGTGGGTTATACCCTACCCACAGCGTCCGGATGTCGGACCGTGCCGAAAGTTCTTTGCAGTCGGAATTAAAGCATTGATCGTTCACAGACGGCGTTCGAGTTAGAAGGCAGCAAGGTTTCGCCAATCCGGAAGGGCTCCCACTATTCAATAGCGTGGCCAGTGTTATTGCTGTTATCACAAAGGCTCTGTTAAACGGTCTAACCCACATCTTGCGGCCAGGTAAAAGCCTCCCGCTCAGAGCCCTAGTTTTGAACTCGAACCGCCCACATATTAATGACTAGAAAGAGCACCGTTAATTTGCAGAGGTTTCAGGTAAATCTATACCCCGGTCAAAACTCGAGTTGTATGGATAAGGTCGGTATATTATCGCGCGTGAAATTAAACCACTGGCGCTCTAATCTGTCCGTCTCTACGGGAGAGAATGCGAATGGTATATTTCAGGATTAGCAAGCCTCCACAGGTGCAGCCAATATAAGGCCCTTGGAAAGAGGGGCGCCCAGTCTCATTCGATAGAACAATTTAAATGTCACAGAATTACCTTGCCGTCCAGAGGCCTGCGAGCGACCCGAAATACCCTTCTTAACACCTCTCAAATCGTGCGTAAAGAACTCTGCGTGCGGCAATGCCCCGGGTCCCTGATTGCTGTTTGCGTACTATCTTTTTTTATTAACGGTCGCTTCATGAACTATCTACGTTTTTTGGCGTGACCTTCTAGGGTTACTCCACAACTAAGTACTTGAGTGTACACGGTGGCCGCCAGGCTTCAGTAGAACTATTCCAACAAGATGAGAGGCTGCGGAGTCCATTCCGACCTGCCTGAACACCATGGCTTTATAGGCTGGACCCGAGTGACCGTCTGCTTTACGCCCTGAACACTAATCACTAGCTGCCCCGGGATTTGTTATATAGAGCGCCTCCCTAGATCCCACCTTAGTCCTATGCACGATACTTATGTTGGAATATATCCCTTAGTGAGGGAAGGTCGACGGGCCGTAAATGTCTTTAACTCCATGACGTTCTATTCCGCAAGGCCGGGCGCATTCATCGCTACACGGATTATTACATTTGTCGTTCCTGGCCCGCTCAGAGATCTAGAGCTAAACGTTCTTGAGGCCAAAATGAGTCCGGCCCTGCAATATGGTTCCGTCGGACCGCTACCCTTCGATCGGTGTCATAAACCTCCGTTAAGCGGAACAGGCTCGGCCACTGACGAAAGCGTCGGTCACAGACCCCTTATCTGAGTAGTATATTCAGTTGGTTTTAGAACACGTACTTCCTGGGCAGGTGCGAGCGGATTTTGATAGTGATAAAGCCCGGCGCGCACTAGGGAACACCTGAATTCAACTGTCTCTGTGTATCGTCCTTTGCTTCCGCAGCCGAATCGTACACTACCTGCCGAGACTGTAAAATAACGGCAAGTAACTTAAAGACCTGGGTGAATCACTTACGAATTACCGTGGTATATTATTTGCTGAAGATGATCTCGCCCCTCGGGTGAGCACCATTAGGTTACACCACACAACATTCGGATGACTTCAACGAGTAGGGGGTACTGAGACGGATCATCACCAGACCTCGTAATAGTACCATTTGTTAGGTCCGCAGGCGAATGCTTACCAACAACACGCAAAAATCGGTGCGCCAATTCATCCAACAGAATTTGATGTCTATGGCATAAAGGCACTCGGATAGTGCCCCTAAGAGATCAGAATCTGTGGAAAAACGTCCGTACTCAGTGCTCGGGCAGTACCGCATCCGAGTGGCCCTATGTTCAGGTGGAGCTCCCGACGAGATCATAACGCTCAGACCTAGCTACCTAGCGGTAACTCCGGTGATATCCACGGCCGCCACACGACTCTCGCTGCAATGGAGCAAGGGGAATTTGGTTCAACGCCGGGTTGAACGAGAACCATTGTGTTTTTCAGGTATGCATTGCGGTCAGCAACTTCAAGGGATTCGTTGGTTGTATGTCTATATCACCTCAAGTCAGGCGTTTAGTGCAAACTTGCAAAGGGCGCGTTCATTCGTTATATGGGTGATCTAACCATTATAAATGTGCCATTTACCAGCCGCACTGAGGGATCTCCAATTTTCCCGCAATTAATTGTTAAGTTGCCTGTTCCGGTACCCTCGTGTATGGACGCTAGCGGACTATTGGTTAAATAAAACCAGTCGTATGCCGCCCAATTAAGCGGGGGCCCCCCAATTCAACAGGCCAGTTGCGCCGGTTCTGCTGCTCCGGCCCCACGCCTCGCTGTGCTAAAGTAGATGCTTAGGTTTTACCGCCACCTGACACGTCATAGCAGAGCCTAGACGGAAAAAACTAAGGTAGACCCTGCAGCTTAGGTGGCGCCTCAGACCGACAGGCACGTAAGCTGCCAGGGCCCTGTTCACGTAGGACGCTACTGACTTCCTTGCGGAACTTGTAGGTTAGACTCAATCGCCGTACGCTCTTTATCTGAGAGCAAAGCACTGAATGACAGCAGACGACTATATTTTTCCTCTGAAGCAACGACCTATTCATAGGAGTAGCGATGCTAAGAAGTGTCATCAGAGAATGGTTGCTTCATCCTTATGGACAGTGATGTTTCACGATCACACCCCTAGCCAGCAGCCCAATATCTATGATTACAATATGGGTCTTCATGATACTTTTTGAGCATGTACAGTGTAGACTCCCTGGACGGTTGAGAACACCACGAGATCATGCGAGCAATGCCACACCACACAAGATTGTAAGCTCCTGTCATGATAGCCTTCTACAATGGGTCTCTG"
complement(string)
```
