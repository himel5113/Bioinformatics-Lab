## Problem: BA2C
Find a Profile-most Probable k-mer in a String
#### link: https://rosalind.info/problems/ba2c/

```python
import math
```

```python
def probability(kmer, matrix):
    ans = 1
    for i in range(len(kmer)):
        if kmer[i] == 'A':
            ans *= matrix[0][i]
        if kmer[i] == 'C':
            ans *= matrix[1][i]
        if kmer[i] == 'G':
            ans *= matrix[2][i]
        if kmer[i] == 'T':
            ans *= matrix[3][i]
    return ans

def most_probable_kmer(dna, k, matrix):
    n = len(dna)
    prob = -1e9
    ans = ""
    for i in range(n - k + 1):
        kmer = dna[i:i + k]
        x = probability(kmer, matrix)
        if x > prob:
            prob = x
            ans = kmer
    return ans
```

```python
profile_matrix = [
[0.32, 0.32, 0.08, 0.24, 0.36, 0.16, 0.24, 0.2],
[0.2, 0.32, 0.24, 0.2, 0.32, 0.24, 0.36, 0.16],
[0.2, 0.16, 0.2, 0.28, 0.28, 0.4, 0.08, 0.36],
[0.28, 0.2, 0.48, 0.28, 0.04, 0.2, 0.32, 0.28],
]
dna = 'CCGTGTAGCCCGACGTCTATGCATCTACCACCTAAAGGTCCAACAGCCTCTACTCATATCGAATGGGCTATGTTCATGCGATTGTTTACTGCTCAGCACCAAGATCAATCGAAGTACCTTCGGGAGGCCTTACACGGCTCCTGAGTGCGGTTAGGTGTTTTTCCAGAACGCCCAACAACACTGCTTAAGTAGAGAGAGAT'
k = 8
most_probable_kmer(dna, k, profile_matrix)
```
