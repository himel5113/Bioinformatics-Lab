## Problem: BA2B
Find a Median String
#### link: https://rosalind.info/problems/ba2b/

```python
import itertools
import sys
import math
```

```python
# all possible k-mers

def generate_kmers(k):
    bases = ['A', 'C', 'G', 'T']
    all_kmers = [''.join(p) for p in itertools.product(bases, repeat = k)]
    return all_kmers
```

```python
generate_kmers(3)
```

```python
def hamming_distance(a, b):
    cnt = 0
    for i in range(len(a)):
        cnt += (a[i] != b[i])
    return cnt

def minimum_hamming_distance(pattern, dna):
    n = len(dna)
    k = len(pattern)
    min_dist = float(math.inf)
    for i in range(n - k + 1):
        kmer = dna[i:i + k]
        cnt = hamming_distance(pattern, kmer)
        if min_dist > cnt:
            min_dist = cnt
    return min_dist
```

```python
# Media String
def median_string(dna_list, k):
    kmers = generate_kmers(k)
    best_pattern = ""
    best_dist = float(math.inf)
    pattern_dist = {}
    for pattern in kmers:
        dist = 0
        for dna in dna_list:
            dist += minimum_hamming_distance(pattern, dna)
        pattern_dist[pattern] = dist
        if dist < best_dist:
            best_dist = dist
            best_pattern = pattern
            
    return best_pattern
```
