## Problem: BA2A
Implement MotifEnumeration
#### link: https://rosalind.info/problems/ba2a/

```python
import itertools
```

```python
# all possible k-mers
def generate_kmers(k):
    bases = ['A', 'C', 'G', 'T']
    all_kmers = [''.join(p) for p in itertools.product(bases, repeat = k)]
    return all_kmers


# Hamming Distance
def hamming_distance(a, b):
    cnt = 0
    for i in range(len(a)):
        cnt += (a[i] != b[i])
    return cnt


# if candidate appears in a DNA string with <= d mismatch
def check(pattern, dna_string, d):
    n = len(dna_string)
    k = len(pattern)
    for i in range(n - k + 1):
        kmer = dna_string[i:i+k]
        if hamming_distance(kmer, pattern) <= d:
            return True
    return False



def MotifEnumeration(k, d, dnaList):
    kmers = generate_kmers(k)
    motif = set()
    for kmer in kmers:
        if all(check(kmer, dna, d) for dna in dnaList):
            motif.add(kmer)
    return motif
```

```python
k, d = 5, 2
dna_list = [
'CGTCACGTTAGCCGAGAGGATTCTA',
'AAAACACCCCCCCCTCGACACTCGC',
'GAAAAGGTGATAGCCCGGCAGGGTG',
'CGTGATCGAATAGGCCGACATCGGT',
'ATCAGAGCTGTTCTCTATTTCGACA',
'CGCGAACAACTCACGCAGAACTGTG',
'TACGAAGGCGTTGTATATTGCGCTA',
'ACGCTCGACAGTACTGGTCACACCA',
'TTTCGCATCAGGTAGCACATCGAAA',
'TCGCGTCTCAACGTCCGACAGAGGT',
]

t = MotifEnumeration(k, d, dna_list)
for i in t:
    print(i, end = ' ')
```
